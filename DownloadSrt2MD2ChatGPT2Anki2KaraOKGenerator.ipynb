````python


# ============================================================
# ğŸµ Anki Karaoke Generator (HTML ä¿ç•™ + éŸ³é¢‘åˆå¹¶ç‰ˆ v7)
# ä½œè€…: ChatGPT (GPT-5)
# ============================================================
from google.colab import drive
drive.mount("/content/drive")

!pip install edge-tts beautifulsoup4 pandas pydub > /dev/null

!sudo apt-get update -y
!sudo apt-get install --reinstall ca-certificates -y
!sudo update-ca-certificates

import os, re, shutil, asyncio, csv
import pandas as pd
from bs4 import BeautifulSoup
import edge_tts

from IPython.display import HTML
from io import StringIO
import random
from pydub import AudioSegment  # ğŸ”Š æ–°å¢ï¼šéŸ³é¢‘åˆå¹¶æ”¯æŒ
#import ssl
#ssl._create_default_https_context = ssl._create_unverified_context



# ============================================================
# ğŸ“‚ 1. æŒ‚è½½ Google Drive å¹¶è®¾ç½®æ–‡ä»¶è·¯å¾„
# ============================================================
#anki_path = "/content/drive/MyDrive/ChatGPTAnkiKaraOK/ChatGPT - AWS_SAA_C03_4.txt"
#anki_path = "/content/drive/MyDrive/ChatGPTAnkiKaraOK/ChatGPT - AWS_SAA_C03_3.txt"
#anki_path = "/content/drive/MyDrive/ChatGPTAnkiKaraOK/ChatGPT - AWS_SAA_C03_2.txt"
anki_path = "/content/drive/MyDrive/ChatGPTAnkiKaraOK/ChatGPT - AWS_SAA_C03_1.txt"

#anki_path = "/content/drive/MyDrive/ChatGPTAnkiKaraOK/AWS_SAA_Exam_cn.txt"
#anki_path = "/content/drive/MyDrive/ChatGPTAnkiKaraOK/AWS_SAA_Exam_en.txt"


#output_dir = "/content/drive/MyDrive/ChatGPTAnkiKaraOK/karaoke_v7"
output_dir = "/content/drive/MyDrive/ChatGPTAnkiKaraOK/SAA_1"
audio_dir = os.path.join(output_dir, "audio")
os.makedirs(audio_dir, exist_ok=True)

# ============================================================
# ğŸ§¹ 2. æ¸…ç†æ—§æ–‡ä»¶å¤¹
# ============================================================
if os.path.exists(output_dir):
    resp = input(f"æ˜¯å¦æ¸…ç©ºæ—§çš„ {output_dir} æ–‡ä»¶å¤¹ï¼Ÿ(y/n): ").lower()
    if resp == "y":
        shutil.rmtree(output_dir)
os.makedirs(output_dir, exist_ok=True)
os.makedirs(audio_dir, exist_ok=True)

# ============================================================
# ğŸ“– 3. è‡ªåŠ¨æ£€æµ‹åˆ†éš”ç¬¦å¹¶è¯»å–æ–‡ä»¶
# ============================================================
with open(anki_path, "r", encoding="utf-8") as f:
    lines = [l for l in f.readlines() if not l.startswith("#") and l.strip() != ""]
"""
sample = "".join(lines[:50])
try:
    dialect = csv.Sniffer().sniff(sample)
    sep = dialect.delimiter
    print(f"âœ… è‡ªåŠ¨æ£€æµ‹åˆ°åˆ†éš”ç¬¦: {repr(sep)}")
except Exception:
    sep = "\t"
    print("âš ï¸ æ— æ³•æ£€æµ‹åˆ†éš”ç¬¦ï¼Œé»˜è®¤ä½¿ç”¨åˆ¶è¡¨ç¬¦(\\t)")
"""
sep = "\t"
data_io = StringIO("".join(lines))
df = pd.read_csv(data_io, sep=sep, dtype=str, engine="python", on_bad_lines="skip").fillna("")

col_index = 1
html_blocks = df.iloc[:, col_index].tolist()

# ============================================================
# ğŸ§© 4. HTML è§£æå‡½æ•°ï¼ˆä¿ç•™æ ¼å¼ + è·³è¿‡å›¾æ ‡ï¼‰
# ============================================================
def remove_emoji(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"
        "\U0001F300-\U0001F5FF"
        "\U0001F680-\U0001F6FF"
        "\U0001F1E0-\U0001F1FF"
        "\U00002700-\U000027BF"
        "\U0001F900-\U0001F9FF"
        "\U00002600-\U000026FF"
        "\U00002B50-\U00002B55"
        "]+",
        flags=re.UNICODE
    )
    text = emoji_pattern.sub("", text)
    text = re.sub(r"[^\u4e00-\u9fa5A-Za-z0-9ï¼Œã€‚ï¼ï¼Ÿã€ï¼›ï¼šâ€œâ€â€˜â€™,.!?~\- ]+", "", text)
    return text.strip()

def extract_lines_from_html(html):
    soup = BeautifulSoup(html, "html.parser")
    for br in soup.find_all("br"):
        br.replace_with("\n")
    text = soup.get_text("\n")
    segments = re.split(r"[ã€‚ï¼ï¼Ÿ!?;\n]+", text)
    lines = []
    for seg in segments:
        seg = seg.strip()
        if not seg:
            continue
        seg_no_emoji = remove_emoji(seg)
        if not re.search(r"[\u4e00-\u9fa5A-Za-z0-9]", seg_no_emoji):
            print(f"âš ï¸ è·³è¿‡æ— æ•ˆå†…å®¹ï¼š{seg[:15]}...")
            continue
        lines.append(seg)
    return lines

# ============================================================
# ğŸ—£ï¸ 5. Edge TTS è¯­éŸ³ç”Ÿæˆ
# ============================================================
async def gen_tts(text, out_path, voice):
    text_no_emoji = remove_emoji(text)
    if not re.search(r"[\u4e00-\u9fa5A-Za-z0-9]", text_no_emoji):
        print(f"âš ï¸ è·³è¿‡æ— æ•ˆå†…å®¹ï¼š{text}")
        return
    for attempt in range(3):
        try:
            tts = edge_tts.Communicate(text_no_emoji, voice)
            await tts.save(out_path)
            return
        except Exception as e:
            print(f"âš ï¸ ç”ŸæˆéŸ³é¢‘å¤±è´¥ï¼Œé‡è¯• {attempt+1}/3: {e}")
            await asyncio.sleep(2)
    print(f"âŒ æ— æ³•ç”ŸæˆéŸ³é¢‘: {text[:30]}...")

# ============================================================
# ğŸ¬ 6. Karaoke HTML æ¨¡æ¿
# ============================================================
template = """<!DOCTYPE html>
<html lang="zh">
<head>
<meta charset="UTF-8">
<title>Karaoke {index}</title>
<style>
body {{
  font-family: sans-serif;
  text-align: center;
  background: #111;
  color: #eee;
  overflow: hidden;
}}
#lyrics {{
  max-height: 70vh;
  overflow-y: auto;
  scroll-behavior: smooth;
}}
.line {{
  margin: 12px 0;
  font-size: 1.4em;
  transition: color 0.3s, transform 0.3s;
}}
.active {{
  color: #00eaff;
  font-weight: bold;
  transform: scale(1.1);
}}
audio {{
  margin-top: 20px;
  width: 80%;
}}
</style>
</head>
<body>
<h2>ç¬¬ {index} é¦–</h2>
<div id="lyrics">
{lyrics}
</div>
<audio id="player" controls preload="auto"></audio>
<script>
const lines = {lines_json};
const player = document.getElementById("player");
const elems = [...document.querySelectorAll('.line')];

async function playSeq() {{
  for (let i = 0; i < lines.length; i++) {{
    elems.forEach(e => e.classList.remove('active'));
    elems[i].classList.add('active');
    player.src = lines[i];
    player.play();
    elems[i].scrollIntoView({{ behavior: 'smooth', block: 'center' }});
    await new Promise(r => player.onended = r);
  }}
}}
document.body.addEventListener('click', playSeq);
</script>
<p style="font-size:0.9em;color:#888;">ç‚¹å‡»é¡µé¢å¼€å§‹æ’­æ”¾ ğŸµ</p>
</body>
</html>"""

# ============================================================
# ğŸš€ 7. å¤„ç†æ¯ä¸ª HTML å— + åˆå¹¶éŸ³é¢‘
# ============================================================
async def process_html_block(html, idx, voice):
    lines = extract_lines_from_html(html)
    if not lines:
        return
    subdir = os.path.join(output_dir, f"{idx:04d}")
    os.makedirs(subdir, exist_ok=True)
    audio_files = []

    for i, text in enumerate(lines, 1):
        mp3_path = os.path.join(subdir, f"{i:03d}.mp3")
        await gen_tts(text, mp3_path, voice)
        audio_files.append(mp3_path)

    # ğŸ”Š åˆå¹¶éŸ³é¢‘
    combined = AudioSegment.empty()
    for f in audio_files:
        if os.path.exists(f):
            combined += AudioSegment.from_file(f)
    combined_path = os.path.join(audio_dir, f"{idx:04d}_merged.mp3")
    combined.export(combined_path, format="mp3")
    print(f"âœ… åˆå¹¶å®Œæˆ: {combined_path}")

    # ğŸµ ç”Ÿæˆ HTML æ­Œè¯æ–‡ä»¶
    lyric_html = "\n".join(f"<div class='line'>{remove_emoji(t)}</div>" for t in lines if t.strip())
    html_content = template.format(
        index=idx,
        lyrics=lyric_html,
        lines_json=str([os.path.basename(f) for f in audio_files]).replace("'", '"')
    )
    with open(os.path.join(subdir, f"{idx:04d}.html"), "w", encoding="utf-8") as f:
        f.write(html_content)

# ============================================================
# ğŸ§  8. ä¸»æ‰§è¡Œé€»è¾‘
# ============================================================
voices = [
    "zh-CN-XiaoxiaoNeural", "zh-CN-YunxiNeural", "zh-CN-YunjianNeural",
    "zh-CN-YunyangNeural", "zh-CN-XiaochenNeural", "zh-CN-XiaohanNeural",
    "zh-CN-XiaomoNeural", "zh-CN-XiaoruiNeural", "zh-CN-XiaoqiuNeural",
    "zh-CN-XiaoshuangNeural", "zh-CN-XiaoyouNeural", "zh-CN-XiaozhenNeural",
    "zh-CN-YunfengNeural", "zh-CN-YunhaoNeural", "zh-CN-YunxiaNeural",
    "zh-CN-YunzeNeural", "zh-CN-XiaorouNeural", "zh-CN-XiaoxuanNeural",
    "zh-CN-XiaoyiNeural", "zh-CN-XiaomengNeural"
]
#voice = random.choice(voices)

import nest_asyncio
nest_asyncio.apply()

await asyncio.gather(*(process_html_block(html, i, random.choice(voices)) for i, html in enumerate(html_blocks, 1)))
````
